### Assignment 2: Function with Nested Default Arguments

Define a function that takes two arguments, a and b, where b is a dictionary with a default value of an empty dictionary. The function should add a new key-value pair to the dictionary and return it. Test the function with different inputs.

```python
def add_dict(a,b=None):
    if b is None:
        b={}
    b[a] = a**2
    return b
print(add_dict(2))
print(add_dict(3,{1:1}))
```

```python
def even_odd(num):
    '''this function mainly worked for find even and odd number'''
    if num%2==0:
        print(f"the {num} is a even")
    else:
        print(f"the {num} is a odd")
even_odd(24)
```

```python
def add(a,b):
    '''this function add values'''
    c = a+b
    return c
add(2,3)
```

```python
#variable length arguments
#positional and keyword arguments

def print_num(*Zobayer):
    for number in Zobayer:
        print(number)
print_num(1,2,3,4,5,6,7,8,"Zobayer")
```

```python
#keywords arguments
def print_details(*args, **kwargs):
    for val in args:
        print(f"Positional:{val}")
    for key, value in kwargs.items():
        print(f"{key}:{value}")

print_details(1,2,3,4,5,6,7,8,"Zobayer", name = 'Zobayer', age = '32', country="Bangladesh")
```

```python
def convert(temp,unit):
    '''this function converts temp between c and f'''
    if unit == "C":
        return temp*9/5 +32
    elif unit =="F":
        return (temp-32)*5/9
    else:
        return None
print(convert(100,"C"))# celsius to fahrenheit
print(convert(100,"F")) # fahrenheit to celsius
```

```python
def is_strong(password):
    """this functions checks strong password or not"""
    if len(password)<8:
        return False
    elif not any(char.isdigit() for char in password):
        return False
    elif not any (char.islower() for char in password):
        return False
    elif not any (char.isupper() for char in password):
        return False
    elif not any(char in '!@#$%&' for char in password):
        return False
    else:
        return True
print(is_strong('nnnnAabc1'))
print(is_strong('Callin123@'))
```
