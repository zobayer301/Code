```python
import pandas as pd
```

```python
#series
data = [1,2,3,4,5]
series = pd.Series(data)
print(series)
```

```python
##create a series from dict element
data = {'a':1, 'b':2,'c':3}
series_dict = pd.Series(data)
print(series_dict)
```

```python
data = [10,20,30]
index=['a','b','c']
pd.Series(data,index = index)
```

```python
###dataframe
#create a dataframe from a dict of list


data ={
    'Name': ["krish", 'john', 'jack'],
    "Age": [25,22,24],
    'City': ['Dhaka', 'Rajshahi', 'Khulna']
    }
df = pd.DataFrame(data)
print(df)
print(type(df))
```
