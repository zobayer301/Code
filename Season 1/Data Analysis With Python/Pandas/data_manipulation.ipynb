#### Data Manipulation and Analysis with Pandas
Data Manipulation and Analysis are key tasks in any data science or data analysis project. Pandas provides a wide range of functions for data manipulation and analysis, making it easier to clean, transform and extract insights from data. In this lesson, we will cover various manipulation and analysis techniques using Pandas

```python
import pandas as pd
```

```python
df = pd.read_csv('data.csv')

#fetch the first 5 rows
df.head(5)
```

```python
df.tail(5)
```

```python
print(df.describe())
df.dtypes
```

```python
### handling missing value
df.isnull().any(axis=1)
```

```python
df.isnull().sum()
```

```python
df.fillna(0)
```

```python
## filling missing values with the mean of the column
df['Calories_fill']=df['Calories'].fillna(df['Calories'].mean())
df
```

```python
df.dtypes
```

```python
##rename column
df = df.rename(columns={'Duration': 'Time'})
df.head(5)
```

```python
## change datatype
df['Pulse_new'] = df['Pulse'].astype(int)
df.head(5)
```

```python
df['New_Maxpulse'] = df['Maxpulse'].apply(lambda x:x**2)
df.head(5)
```

```python
##data aggregating and grouping
grouped_mean = df.groupby('Time')['Pulse'].mean()
print(grouped_mean)
```

```python
grouped_sum = df.groupby(['Time', 'Pulse']) ['Calories'].sum()
print(grouped_sum)
```

```python
grouped_agg = df.groupby('Time')['Pulse'].agg(['mean','sum','count'])
print(grouped_agg)
```

```python
## merging and joining dataframes
#create sample data frame

df1 = pd.DataFrame({'Key':['A','B','C'], 'Value1':[1,2,3]})
df2 = pd.DataFrame({'Key':['A','B','F'], 'Values2':[4,5,6]})
print(df1)
print(df2)
```

```python
#merge dataframe on the key columns
pd.merge(df1,df2, on='Key', how = 'outer')
```

```python
pd.merge(df1,df2, on='Key', how = 'inner')
```

```python
pd.merge(df1,df2, on='Key', how = 'left')
```

```python
pd.merge(df1,df2, on='Key', how = 'right')
```
