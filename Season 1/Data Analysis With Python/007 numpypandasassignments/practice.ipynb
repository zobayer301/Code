```python
array = np.random.randint(1,21 , size=(5,5))
print(array)

mean = np.mean(array)
median = np.median(array)
variance = np.var(array)
stn_div = np.std(array)

print("Mean:\n ", mean)
print("Median:\n", median)
print("Variance:\n", variance)
print("Standard Deviation:\n", stn_div)
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.randint(1,20, size = (5,3)), columns=['A','B','C'])
print("Original Dataframe")
print(df)

df.iloc[0,1]=np.nan
df.iloc[1,2]=np.nan
df.iloc[4,1]=np.nan
print("After NaN")
print(df)

df.fillna(df.mean(),inplace= True)
print("Filled with mean Values")
print(df)
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Training data (Area in square feet)
X = np.array([[1000],
              [1500],
              [2000],
              [2500],
              [3000]])

# House prices
y = np.array([200000,
              300000,
              400000,
              500000,
              600000])

# Create and train the model
model = LinearRegression()
model.fit(X, y)

# Predict price for a 2200 sq ft house
new_area = np.array([[2200]])
predicted_price = model.predict(new_area)

print("Predicted price:", predicted_price[0])

# Plot the data and regression line
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, model.predict(X), color='red', label='Regression Line')
plt.xlabel('Area (sq ft)')
plt.ylabel('Price')
plt.title('Linear Regression Example')
plt.legend()
plt.show()
```
