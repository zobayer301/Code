### File Operation- Read and Write Files
##### file handling is a crucial part of any programming language. python provides built-in functions and methods to read from and write to files, both text and binary. this lesson will cover the basics of the file handling, including reading and writing text files and binary files.

```python
## read the whole file
with open('example.txt', 'r') as file:
    content = file.read()
    print(content)
```

```python
## read file line by line
with open('example.txt','r') as file:
    for line in file:
        print(line.strip()) ##strip() removes the newline character
```

```python
## writing a file(overwriting)
with open('example.txt', 'w') as file:
    file.write('hello world\n')
    file.write("this is new line")
    
```

```python
#write a file without overwriting
with open('example.txt', 'a') as file:
    file.write("hehehe\n")
```

```python
#write a list of lines to a file
lines = ['fist line\n','second line\n','third line\n']
with open('example.txt','a') as file:
    file.writelines(lines)
```

```python
## binary files
## writing to a binary file
data= b'\x00\x01\x02\x03\x04'
with open('example.bin','wb') as file:
    file.write(data)
```

```python
#reading a binary file
with open('example.bin','rb') as file:
    content = file.read()
    print(content)
```

```python
# read the content from source text file and write to a destination text file
with open('example.txt','r') as source_file:
    content = source_file.read()

with open('destination.txt', 'w') as destination_file:
    destination_file.write(content)
```

```python
# read a file and count the number of lines, words and characters.
def count_text_file(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()
        line_count = len(lines)
        word_count= sum(len(line.split()) for line in lines)
        char_count = sum(len(line)for line in lines)
    return line_count, word_count, char_count
file_path = 'example.txt'
lines, words, characters= count_text_file(file_path)
print(f'Lines:{lines}, Words:{words}, characters: {characters}')
```

The w+ mode in python is used to open a file for both reading and writing, if the file does not exist, it will be created. if the file exits, its content is truncated 

```python
## writing and then reading a file
with open('example.txt', 'w+') as file:
    file.write('hello world\n')
    file.write('this is new line \n')
# move the file cursor to the beginning
    file.seek(0)
    
# read the content of the file
    content=file.read()
    print(content)
```
