### Iterators

Iterators are advanced Python concepts tht allow for efficient looping and memory management. Iterators provide a way to access elements of a collection sequentially without exposing the underlying structure.

```python
my_list=[1,2,3,4,5,6]

for i in my_list:
    print(i)
```

```python
##iterators
iterator=iter(my_list)
print(type(iterator))
```

```python
next(iterator)
```

```python
iterator=iter(my_list)
```

```python

try:
    print(next(iterator))
except StopIteration:
    print("There are no elements in the iterator")
```

```python
#string iterator
my_string = "Hello"
string_iterator = iter(my_string)
print(next(string_iterator))
print(next(string_iterator))
print(next(string_iterator))
print(next(string_iterator))
print(next(string_iterator))
```
