#### Decorators
Decorators are a powerful and flexible in Python that allows you to modify the behavior of a function or class method. They are commonly used to add functionality to functions or methods without modifying their actual code. This lesson covers the basics of decorators, including how to create and use them.

```python
##function copy
## closures
## decorators
```

```python
##function copy
def welcome():
    return 'Welcome to the course'
welcome()
```

```python
wel=welcome
print(wel())
del welcome
print(wel())
```

```python
## closures functions
def main_welcome():
    msg='Welcome'
    def sub_welcome_method():
        print("Welcome to the python course")
        print(msg)
        print("please learn these concept properly")
    return sub_welcome_method()
```

```python
main_welcome()
```

```python
#decorator
def main_welcome(func):
    msg='Welcome'
    def sub_welcome_method():
        print("Welcome to the python course")
        func()
        print("please learn these concepts properly")
    return sub_welcome_method()
```

```python
def course_intro():
    print("This is an advanced python course")
course_intro()
```

```python
main_welcome(course_intro)
```

```python
@main_welcome
def course_intro():
    print("This is an advanced python course")
```

```python
##decorator
def my_decorator(func):
    def wrapper():
        print("Something is happening before the function is called")
        func()
        print('Something is happening after the function is called')
    return wrapper
```

```python
@my_decorator
def say_hello():
    print("Hello")
```

```python
say_hello()
```

```python
## decorators with arguments
def repeat(n):
    def decorator(func):
        def wrapper(*args, **kwargs):
            for _ in range(n):
                func(*args, **kwargs)
        return wrapper
    return decorator
```

```python
@repeat(2)
def say():
    print("Hello")
```

```python
say()
```
