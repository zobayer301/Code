# Tuple

tuples are ordered collections of items that are immutable. they are similar to lists, but their immutability makes them different

```python
#creating a tuple
empty=()
print(empty)
print(type(empty))
```

```python
lst= list()
print(type(lst))
tpl = tuple()
print(type(tpl))
```

```python
numberr = tuple([1,2,3,4])
numberr
```

```python
mixed = (1, "hello", 3.12, True)
print(mixed)
```

```python
#accessing tuple
print(numberr[2])
print(numberr[-1])
```

```python
#tuple operations
concatenation = numberr + mixed
print(concatenation)
```

```python
mixed*2
```

```python
# immutable nature of tuple
numberr[1] = "zobayer"
```

```python
#tuple method
print(numberr.count(1))
print(numberr.index(3))
```

```python
#packing and unpacking tuple
packed_tuple = 1, "hello", 3.14
print(packed_tuple)
```

```python
#unpacking tuple
a,b,c, = packed_tuple
print(a)
print(b)
print(c)
```

```python
#unpacking with *
numberss = (1,2,3,4,5,6)
first, *middle,last = numberss
print(first)
print(*middle)
print(last)
```

```python
#nested list
lst = [[1,2,3,4,5,6],[7,8,9],[1,"hello","c"]]
print(lst[0][3])
print(lst[1])
print(lst[2])
```

```python
#nested tuple
lst_tup = ((1,2,3,4,5,6),(7,8,9),(1,"hello","c"))
print(lst_tup[0][3])
print(lst_tup[1])
print(lst_tup[2])
```

```python
#iterating over nested tuple
for sub_tup in lst_tup:
    for item in sub_tup:
        print(item, end="")
    print()
```

# Sets

Sets are a built-in data type in python used to store collection of unique items. They are unordered, meaning the elements do not follow a specific order

```python
#create a set
my_set = {1,2,3,4}
print(my_set)
print(type(my_set))
```

```python
emty_set = set()
print(type(emty_set))
```

```python
# basic set operation
#adding and removing elements
my_set.add(7)
print(my_set)

my_set.add(7)
print(my_set)
```

```python
my_set.remove(3)
print(my_set)
```

```python
my_set = {1,2,3,4}
removed_ele = my_set.pop()
print(removed_ele)
print(my_set)
```

```python
set1 = {1,2,3,4,5,6}
set2 = {4,5,6,7,8,9}

#union
union_set = set1.union(set2)
print(union_set)

#intersection
intersection_set = set1.intersection(set2)
print(intersection_set)

```

```python
#set method
set3 = {1,2,3}
set4 = {3,5,6}
#is subset
print(set3.issubset(set4))
print(set3.issuperset(set4))
```

# Dictionaries

#####  Dictionaries are unordered collection of items.  they store data in key-value pairs. Keys must be unique and immutable while values can be of any type.

```python
student = {"name": "Zobayer", "age":"23"}
print(student)
```

```python
#error
student = {"name": "Zobayer", "age":"23", "name":"istiak"}
print(student)
```

```python
#accessing dict elements
student = {"name": "Zobayer", "age":"23","grade":"A"}
print(student['grade'])
print(student['age'])

# accessing using get() method
print(student.get('grade'))
print(student.get('last_name', "Not Available"))
```

```python
## Modifying Dictionary elements
#dic are mutable, so you can update or delete elements
print(student)
student['age'] = 32 # update the value
student['address'] = "Bangladesh" # added a new key and value
print(student)

del student['grade'] # delete key and value pair
print(student)
```

```python
#dictionary methods
keys = student.keys()
print(keys)
values = student.values()
print(values)

item = student.items() #get all key values and pairs
print(item)
```

```python
student_copy1 = student.copy() #shallow copy
student["name"]= "Zobayer"

print(student_copy1)
print(student)
```

```python
# iterating over dict
#loops to iterate over dict, keys, values or items
for keys in student.keys():
    print(keys)
```

```python
for values in student.values():
    print(values)
```

```python
for items in student.items():
    print(items)
```

```python
#nested dict

students ={
   "student1" : {"name":"Zobayer", 'age':25},
   "student2" : {"name": "Chowdhury", "age":23}
}

print(students)
```

```python
## access nested dict
print(students['student2']['name'])
print(students['student2']['age'])
```

```python
#dictionary comprehension
squares = {x: x**2 for x in range(5)}
print(squares)
```

```python
#conditional dict comprehension
even ={ x:x**2 for x in range(10) if x%2==0}
print(even)
```

```python
#practical examples
# use a dictionary to count the frequency of elements in list
numbr = [1,2,2,3,3,3,4,4,4,4]
frequency = {}

for number in numbr:
    if number in frequency:
        frequency[number]+=1
    else:
        frequency[number]=1
print(frequency)
```
