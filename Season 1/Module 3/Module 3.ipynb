```python
age = 20

if age<13:
    print("you are child")
elif age<18:
    print("you are teenager")
else:
    print("you are adult")
```

```python
#nested conditional statements
num = int(input("Enter any Number: "))

if num>0:
    print('The number is Positive')
    if num%2==0:
        print("The number is EVEN")
    else:
        print('The number is ODD')
elif num == 0:
    print("The number is ZERO")
else:
    print("the number is NEGATIVE")
```

```python
#leap year
ye = int(input("Enter any year: "))
if ye%4==0:
    if ye%100==0:
        if ye%400==0:
            print("leap")
        else:
            print("Not leap")
    else:
        print("leap")
else:
    print("Not leap")
```

# Loops

```python
range(0,5)
```

```python
for i in range(0,5):
    print(i)
```

```python
for i in range(10,1,-2):
    print(i)
```

```python
#string
str = 'Zobayer'
for i in str:
    print(i)
```

```python
#while
count = 0
while count<5:
    print(count)
    count = count+1
```

```python
count = 0
while count%2==0:
    print(count)
    count = count+1
```

```python
#loop control statements
#break
for i in range(10):
    if i==5:
        break
    print(i)
```

```python
#continue
for i in range(10):
    if i % 2==0:
      continue
    print(i)
```

```python
##pass
# the pass statement is a null, it does nothing
for i in range(5):
    if i==3:
        pass
    print(i)
```

```python
#nested loops
for i in range(3):
    for j in range(2):
        print(f"i:{i} and j:{j}")
```

```python
#examples - calculate the sum of first N natural numbers using a while and for loop
# while loop
n=50
sum = 0
count=1

while count<=n:
    sum = sum+count
    count = count+1
print("sum of first 10 natural number: ", sum)
```

```python
n=10
sum = 0
for i in range (5):
    sum = sum*i
print(sum)
```

```python
#example - prime numbers between 1 to 100
for num in range(1,101):
    if num>1:
        for i in range (2,num):
            if num%i==0:
                break
        else:
                print(num)
```
