```python
def divide(a,b):
    try:
        result = a/b
    except ZeroDivisionError as e:
        print(f"Error: {e}")
        result = None
    return result
print(divide(4,2))
print(divide(4,0))
```

```python
def read_file(filename):
    try:
        file = open(filename, 'r')
        content = file.read()
        return content
    except FileNotFoundError as e:
        print(f"Error: {e}")
    finally:
        try:
            file.close()
        except NameError:
            pass
print(read_file('data.txt'))
```
