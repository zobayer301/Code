## Handling Imbalanced Dataset
1. Up Sampling
2. Down Sampling

```python
import pandas as pd
import numpy as np

np.random.seed(123)

n_samples = 1000
class_0_ratio = 0.9

n_class_0 = int(n_samples * class_0_ratio)
n_class_1 = n_samples - n_class_0
```

```python
n_class_0, n_class_1
```

```python
class_0 = pd.DataFrame({
    'feature1': np.random.normal(loc=0, scale=1, size=n_class_0),
    'feature2': np.random.normal(loc=0, scale=1, size=n_class_0),
    'target': [0] * n_class_0
})

class_1 = pd.DataFrame({
    'feature1': np.random.normal(loc=2, scale=1, size=n_class_1 ),
    'feature2': np.random.normal(loc=2, scale=1, size=n_class_1),
    'target': [1] * n_class_1
})

```

```python
df= pd.concat([class_0, class_1]).reset_index(drop=True)
```

```python
df
```

```python
df['target'].value_counts()
```

```python
df_majority = df[df['target']==0]
df_minority = df[df['target']==1]
```

```python
from sklearn.utils import resample
df_minority_upsample = resample(df_minority, replace=True, n_samples=len(df_majority), random_state=42)
```

```python
df_minority_upsample.shape
```

```python
df_upsample = pd.concat([df_majority, df_minority_upsample])
```

```python
df_upsample['target'].value_counts()
```

### Down Sampling

```python
from sklearn.utils import resample
df_majority_downsample = resample(df_majority, replace=False, n_samples=len(df_minority), random_state=42)
```

```python
df_downsample = pd.concat([df_minority, df_majority_downsample])
```

```python
df_downsample['target'].value_counts()
```
